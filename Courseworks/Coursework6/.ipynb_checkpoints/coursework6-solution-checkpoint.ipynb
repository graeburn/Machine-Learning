We write two functions **standardise** and **de_standardise** to (de-)standardise the columns of a two-dimensional array. The function **standardise** takes the two-dimensional array *matrix* as its input argument. It subtracts the means from each column and divides by the standard deviations. It returns the standardised matrix, the row of means and the row of standard deviations. The function **de\_standardise** reverses this operation. It takes a standardised matrix, the row of means and the row of standard deviations as its arguments and returns a matrix for which the standardisation process is reversed.

```python
def standardise(matrix):
    row_of_means = np.mean(matrix, axis=0)
    standardised_matrix = matrix - row_of_means
    row_of_stds = np.std(standardised_matrix, axis=0)
    return (standardised_matrix / row_of_stds), row_of_means, row_of_stds

def de_standardise(standardised_matrix, row_of_means, row_of_stds):
    matrix = np.copy(standardised_matrix * row_of_stds)
    return matrix + row_of_means
```

We write two functions **mean_squared_error** and **mean_squared_error_gradient** that implement the mean squared error and its gradient as defined in the lecture notes. Both functions take a two-dimensional NumPy array *data_matrix*, a one-dimensional NumPy array *weights* and a one-dimensional NumPy array *outputs* as arguments.

```python
def mean_squared_error(data_matrix, weights, outputs):
    return np.mean((data_matrix @ weights - outputs) ** 2) / 2

def mean_squared_error_gradient(data_matrix, weights, outputs):
    return (data_matrix.T @ (data_matrix @ weights - outputs)) / len(outputs)
```

We compute the value *optimal_mean_squared_error* of the mean-squared-error for the optimal weights *optimal_weights* that we have computed earlier.

```python
optimal_mean_squared_error = mean_squared_error(data_matrix, optimal_weights, data_samples)
print("The optimal mean squared error is {mse}.".format(mse=optimal_mean_squared_error)) #Reference value
```

Next, we implement a function **gradient_descent** that performs gradient descent to numerically approximate a minimiser of a convex function *objective* with gradient *gradient*. In addition to the functions *objective* and *gradient* that take a NumPy array as their argument, the arguments for this function are a NumPy array *initial_weights* with initial values for the first iterate, a step-size parameter *step_size* for the gradient descent step and a parameter *no_of_iterations* that controls the number of iterations. We implement the function so that it returns a NumPy array of the weights obtained after gradient descent together with a list of objective values *objective_values* for all iterates.

```python
#Print every 10 iterations to check working. Need to make COPY of initial weights so don't update them
def gradient_descent(objective, gradient, initial_weights, step_size=1, \
                     no_of_iterations=100, print_output=10):
    objective_values = []
    weights = np.copy(initial_weights)
    objective_values.append(objective(weights))
    for counter in range(no_of_iterations):
        weights -= step_size * gradient(weights) #Implement gradient descent
        objective_values.append(objective(weights))
        if (counter + 1) % print_output == 0:
            print("Iteration {k}/{m}, objective = {o}.".format(k=counter+1, \
                    m=no_of_iterations, o=objective_values[counter]))
    print("Iteration completed after {k}/{m}, objective = {o}.".format(k=counter + 1, \
                m=no_of_iterations, o=objective_values[counter]))
    return weights, objective_values
```

We make use of *mean_squared_error*, *mean_squared_error_gradient* and Python's **lambda**-function to set up the arguments *objective* and *gradient*.

```python
# Lambda allows use of the gradient descent function with the mse function (defined with its own set of arguments)
objective = lambda weights: mean_squared_error(data_matrix, weights, data_samples)
gradient = lambda weights: mean_squared_error_gradient(data_matrix, weights, data_samples)
```

We use the function **gradient_descent** to minimise the mean squared error (and thus, to compute a solution *gradient_descent_weights* to the linear regression problem) with zeros as initial weights.

```python
initial_weights = np.zeros((data_matrix.shape[1], 1)) #Zeros in right shape as matrix
no_of_iterations=100
step_size = (1.9 * len(data_samples))/(np.linalg.norm(data_matrix, 2) ** 2) #Verified in notes for mse. Other functions try values or use backtracking modification
gradient_descent_weights, objective_values = gradient_descent(objective, gradient, \
                                    initial_weights, step_size, no_of_iterations, 10)
```

```python
# Reproduced similar solution by gradient descent and 100 iterations
print("The weights obtained with gradient descent after {it} iterations".format( \
        it=no_of_iterations),"\nand step-size {s} are {w}.T".format(s=step_size, \
        w= gradient_descent_weights.T))
```

## Proximal gradient descent

In this exercise we extend the concept of gradient descent to proximal gradient descent. Subsequently, we use proximal gradient descent to implement the LASSO problem introduced in the lecture. Based on our function **gradient_descent** from earlier we write a function **proximal_gradient_descent** that takes all the arguments of **gradient_descent** but also a function *proximal_map*.

```python
def proximal_gradient_descent(objective, gradient, proximal_map, initial_weights, \
                              step_size=1, no_of_iterations=100, print_output=10):    
    objective_values = []
    weights = initial_weights    
    objective_values.append(objective(weights))
    for counter in range(no_of_iterations):
        weights = proximal_map(weights - step_size * gradient(weights)) #Now with extra proximal map
        objective_values.append(objective(weights))
        if (counter + 1) % print_output == 0:
            print("Iteration {k}/{m}, objective = {o}.".format(k=counter+1, \
                    m=no_of_iterations, o=objective_values[counter]))
    print("Iteration completed after {k}/{m}, objective = {o}.".format(k=counter + 1, \
                m=no_of_iterations, o=objective_values[counter]))
    return weights, objective_values
```

We implement the proximal map for the one-norm, also known as **soft_thresholding**. Our function takes a NumPy array *argument* and a scalar *threshold* as its arguments and performs the soft-thresholding operation to each component of the array.

```python
# Proximal map for lasso
def soft_thresholding(argument, threshold):
    return np.sign(argument) * np.maximum(0, np.abs(argument) - threshold)
```

Next, we write a function **lasso_cost_function** that implements the LASSO cost function as introduced in the lecture notes. The arguments include *data_matrix*, *weights* and *outputs* and a positive scalar *regularisation_parameter* that controls the balance between mean squared error and the one norm.

```python
# mse plus regularisation
def lasso_cost_function(data_matrix, weights, outputs, regularisation_parameter):
    return mean_squared_error(data_matrix, weights, outputs) + regularisation_parameter \
            * np.sum(np.abs(weights))
```

Subsequently, we implement the LASSO with the help of the functions implemented previously. We apply the LASSO to the height-weight dataset as well as the sine-function example from [Coursework 5](https://qmplus.qmul.ac.uk/pluginfile.php/2317872/mod_resource/content/2/coursework05.pdf).

### LASSO for the height-weight-dataset

We set the degree *degree* of the poylinomial to $d = 5$ and set up the data matrix that we name *lasso_matrix*.

```python
degree = 5
lasso_matrix = polynomial_basis(height_standardised, degree)
```

Next, we initialise *objective*, *gradient* and *proximal_map*.

```python
regularisation_parameter = 0.5
step_size = (1.9 * len(data_samples))/(np.linalg.norm(lasso_matrix, 2) ** 2)
objective = lambda weights: lasso_cost_function(lasso_matrix, weights, data_samples, \
                                                regularisation_parameter)
gradient = lambda weights: mean_squared_error_gradient(lasso_matrix, weights, data_samples)
proximal_map = lambda weights: soft_thresholding(weights, regularisation_parameter * \
                                                 step_size)
```

Subsequently, we run proximal gradient descent to solve the LASSO problem with regularisation parameter *regularisation_parameter* set to $\alpha = 0.5$.

```python
initial_weights = np.zeros((lasso_matrix.shape[1], 1))
lasso_weights, lasso_objective_values = proximal_gradient_descent(objective, gradient, \
        proximal_map, initial_weights, step_size, no_of_iterations=1000, \
        print_output=100)
```

```python
# Sparse solution as expected
print("The weights obtained with the LASSO are {w}.T.".format(w=lasso_weights.T))
```

We visualise the model applied to the weights that we have obtained.

```python
model_output = de_standardise(lasso_matrix @ lasso_weights, row_of_means, row_of_stds)
plt.plot(height, weight, '.', linewidth=3, label='Samples')
plt.plot(height, model_output, linewidth=3, label='LASSO approximation')
plt.xlabel('Height in metres', fontsize=16)
plt.xticks(fontsize=16)
plt.ylabel('Weight in kilogram', fontsize=16)
plt.yticks(fontsize=16)
plt.legend()
plt.tight_layout;
```

Why did we not obtain a line? You find the answer if you evulate the LASSO cost function for both sets of weights.

```python
# Actually a better fit
value1 = objective(lasso_weights)
value2 = objective(np.r_[optimal_weights, np.zeros((4, 1))])
print("The LASSO objective for the approximation above is {obj}.\nThe".format(obj = value1), \
      "LASSO objective for the line approximation is {obj}.".format(obj = value2))
```

### LASSO for the sine-curve example from Coursework 5

We repeat the previous exercise, but this time we apply the LASSO model to the sine function example from coursework 5. We initialise the samples in exactly the same fashion as on Coursework 5.

```python
no_samples = 1000
sigma = 0.1
inputs = np.linspace(0, 1, no_samples)
true_outputs = np.sin(2*np.pi*inputs)
outputs = true_outputs + sigma * np.random.randn(no_samples)
```

We set the degree *degree* of the poylinomial to $d = 20$ and set up the data matrix that we name *lasso_matrix*.

```python
degree = 20
lasso_matrix = polynomial_basis(inputs.reshape(-1, 1), degree)
```

Next, we initialise *objective*, *gradient* and *proximal_map*.

```python
regularisation_parameter = 5e-4
step_size = 200/(np.linalg.norm(lasso_matrix, 2) ** 2)
objective = lambda weights: lasso_cost_function(lasso_matrix, weights, outputs.reshape(-1, 1), \
                                                regularisation_parameter)
gradient = lambda weights: mean_squared_error_gradient(lasso_matrix, weights, outputs.reshape(-1, 1))
proximal_map = lambda weights: soft_thresholding(weights, regularisation_parameter * \
                                                 step_size)
```

Subsequently, we run proximal gradient descent to solve the LASSO problem with regularisation parameter *regularisation_parameter* set to $\alpha = 5 \times 10^{-4}$.

```python
# Lot of iterations to converge
initial_weights = np.zeros((lasso_matrix.shape[1], 1))
lasso_weights, lasso_objective_values = proximal_gradient_descent(objective, gradient, \
        proximal_map, initial_weights, step_size, no_of_iterations=300000, \
        print_output=30000)
```

We visualise the model applied to the weights that we have obtained.

```python
lasso_outputs = lasso_matrix @ lasso_weights
plt.plot(inputs, outputs, '.-', color='red', label='Noisy function')
plt.plot(inputs, true_outputs, linewidth=3, color='green', label='True function $f$')
plt.plot(inputs, lasso_outputs, ':', linewidth=3, color='blue', label='LASSO reconstruction')
plt.legend( prop={'size': 11})
plt.xlabel('$x$-axis', fontsize=16)
plt.xticks(fontsize=16)
plt.ylabel('$y$-axis', fontsize=16)
plt.yticks(fontsize=16)
plt.tight_layout;
```

```python
# Again sparse though sparsity varies with alpha. Both still better than high order polynomial
print("The weights obtained with the LASSO are {w}.T.".format(w=lasso_weights.T))
```
