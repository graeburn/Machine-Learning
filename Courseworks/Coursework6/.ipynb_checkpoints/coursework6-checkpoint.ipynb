```python
data_samples = np.column_stack((height, weight))
data_samples_sorted = data_samples[data_samples[:,0].argsort()]

print(data_samples_sorted)
```

```python
def standardise(matrix):
    matrix_means = np.mean(matrix, axis=0)
    matrix_stds = np.std(matrix, axis=0)
    standardised = (matrix - matrix_means) / matrix_stds
    
    return standardised, matrix_means, matrix_stds
```

```python
standardised, matrix_means, matrix_stds = standardise(data_samples_sorted)
```

```python
def de_standardise(standardised, matrix_means, matrix_stds):
    matrix = (standardised * matrix_stds) + matrix_means
    
    return matrix
```

```python
matrix = de_standardise(standardised, matrix_means, matrix_stds)
print(matrix)
```

```python
basis_matrix = functions.polynomial_basis(standardised[:,0], 1)
optimal_weights = functions.regression(basis_matrix, standardised[:,1])

model_outputs = basis_matrix @ optimal_weights

output_matrix = de_standardise(np.column_stack((standardised[:,0], model_outputs)), matrix_means, matrix_stds)
```

```python
plt.scatter(height, weight, marker=".", color='blue', label='True function $f$')
plt.scatter(output_matrix[:,0], output_matrix[:,1], marker=".", color='red', label='Best d=1 regression model $f_d$')
plt.legend( prop={'size': 11})
plt.xlabel('height', fontsize=16)
plt.xticks(fontsize=16)
plt.ylabel('weight', fontsize=16)
plt.yticks(fontsize=16)
plt.tight_layout;
```

```python
def mean_squared_error(data_matrix, weights, outputs):
    mse = np.sum((data_matrix @ weights - outputs)**2) / 2*outputs.size
    
    return mse  
```

```python
def mean_squared_error_gradient(data_matrix, weights, outputs): 
    mse_grad = data_matrix.T @ (data_matrix @ weights - outputs) / outputs.size
    
    return mse_grad
```

```python
print(mean_squared_error(basis_matrix, optimal_weights, standardised[:,1]))
print(mean_squared_error_gradient(basis_matrix, optimal_weights, standardised[:,1]))
```

```python
def gradient_descent(objective, gradient, initial_weights, step_size, no_of_iterations):
    next_weights = initial_weights
    objective_values = np.empty(shape=[no_of_iterations+1])
    objective_values[0] = objective(next_weights)
    
    for counter in range(no_of_iterations):
        next_weights = next_weights - step_size*gradient(next_weights)
        objective_values[counter+1] = objective(next_weights)
        
    return next_weights, objective_values
```

```python
weights_0 = [1,1]
my_step_size = 1
my_iterations = 10
obj = lambda x : mean_squared_error(basis_matrix, x, standardised[:,1])
grad = lambda x : mean_squared_error_gradient(basis_matrix, x, standardised[:,1])

linear_reg_weights, linear_reg_mse_values = gradient_descent(obj, grad, weights_0, my_step_size, my_iterations)

print(linear_reg_weights)
print(linear_reg_mse_values)
```

```python
def proximal_gradient_descent(objective, gradient, initial_weights, step_size, no_of_iterations, proximal_map):
    next_weights = initial_weights
    objective_values = np.empty(shape=[no_of_iterations+1])
    objective_values[0] = objective(next_weights)
    
    for counter in range(no_of_iterations):
        next_weights = proximal_map(next_weights - step_size*gradient(next_weights))
        objective_values[counter+1] = objective(next_weights)
        
    return next_weights, objective_values
```

```python
def soft_thresholding(argument, threshold):
    soft_threshold_array = np.empty(shape=argument.size)
    
    for i in range(argument.size):
        if(argument[i] > threshold):
            soft_threshold_array[i] = argument[i] - threshold
        elif(argument[i] < -threshold):
            soft_threshold_array[i] = argument[i] + threshold
        else:
            soft_threshold_array[i] = 0
    
    return soft_threshold_array
    
```

```python
def lasso_cost_function(data_matrix, weights, outputs, regularisation_parameter):
    lasso_cost = np.sum(((data_matrix @ weights - outputs)**2)/2 + regularisation_parameter*np.sum(np.abs(weights)))
    
    return lasso_cost
```

```python
weights_0 = [1,1]
my_step_size = 0.1
my_iterations = 10
regularisation_parameter = 0.5

obj_lasso = lambda x : lasso_cost_function(basis_matrix, x, standardised[:,1], regularisation_parameter)
grad_mse = lambda x : mean_squared_error_gradient(basis_matrix, x, standardised[:,1])
my_map = lambda x : soft_thresholding(x, my_step_size)

my_weights, my_obj_values = proximal_gradient_descent(obj_lasso, grad_mse, weights_0, my_step_size, my_iterations, my_map)

print(my_weights)
print(my_obj_values)
```
