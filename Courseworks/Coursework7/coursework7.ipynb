```python
from collections import Counter

def nearest_neighbour_classification(inputs, dataset, no_of_neighbours):
    predicted_labels = []
    Euclidean_distances = []
    for to_predict in inputs:  
        for point in dataset:
            euclidean_distance_point = np.linalg.norm(np.array(point[:len(point)-1]) - np.array(to_predict))
            Euclidean_distances.append([euclidean_distance_point, point[len(point)-1]])
            
        nearby_label_count = [i[1] for i in sorted(Euclidean_distances)[:no_of_neighbours]]
        result = Counter(nearby_label_count).most_common(1)[0][0]
        predicted_labels.append(result)
    return predicted_labels
```

We write two functions **mean_squared_error** and **mean_squared_error_gradient** that implement the mean squared error and its gradient as defined in the lecture notes. Both functions take a two-dimensional NumPy array *data_matrix*, a one-dimensional NumPy array *weights* and a one-dimensional NumPy array *outputs* as arguments.

```python
def mean_squared_error(data_matrix, weights, outputs):
    return np.mean((data_matrix @ weights - outputs) ** 2) / 2

def mean_squared_error_gradient(data_matrix, weights, outputs):
    return (data_matrix.T @ (data_matrix @ weights - outputs)) / len(outputs)
```

Next, we implement a function **gradient_descent** that performs gradient descent to numerically approximate a minimiser of a convex function *objective* with gradient *gradient*. In addition to the functions *objective* and *gradient* that take a NumPy array as their argument, the arguments for this function are a NumPy array *initial_weights* with initial values for the first iterate, a step-size parameter *step_size* for the gradient descent step and a parameter *no_of_iterations* that controls the number of iterations. We implement the function so that it returns a NumPy array of the weights obtained after gradient descent together with a list of objective values *objective_values* for all iterates.

```python
#Print every 10 iterations to check working. Need to make COPY of initial weights so don't update them
def gradient_descent(objective, gradient, initial_weights, step_size=1, \
                     no_of_iterations=100, print_output=10):
    objective_values = []
    weights = np.copy(initial_weights)
    objective_values.append(objective(weights))
    for counter in range(no_of_iterations):
        weights -= step_size * gradient(weights) #Implement gradient descent
        objective_values.append(objective(weights))
        if (counter + 1) % print_output == 0:
            print("Iteration {k}/{m}, objective = {o}.".format(k=counter+1, \
                    m=no_of_iterations, o=objective_values[counter]))
    print("Iteration completed after {k}/{m}, objective = {o}.".format(k=counter + 1, \
                m=no_of_iterations, o=objective_values[counter]))
    return weights, objective_values
```

We make use of *mean_squared_error*, *mean_squared_error_gradient* and Python's **lambda**-function to set up the arguments *objective* and *gradient*.

```python
# Lambda allows use of same gradient descent function for mse which depends on its own set of arguments
objective = lambda weights: mean_squared_error(data_matrix, weights, data_samples)
gradient = lambda weights: mean_squared_error_gradient(data_matrix, weights, data_samples)
```

## Proximal gradient descent

In this exercise we extend the concept of gradient descent to proximal gradient descent. Subsequently, we use proximal gradient descent to implement the LASSO problem introduced in the lecture. Based on our function **gradient_descent** from earlier we write a function **proximal_gradient_descent** that takes all the arguments of **gradient_descent** but also a function *proximal_map*.

```python
def proximal_gradient_descent(objective, gradient, proximal_map, initial_weights, \
                              step_size=1, no_of_iterations=100, print_output=10):    
    objective_values = []
    weights = initial_weights    
    objective_values.append(objective(weights))
    for counter in range(no_of_iterations):
        weights = proximal_map(weights - step_size * gradient(weights)) #Now with extra proximal map
        objective_values.append(objective(weights))
        if (counter + 1) % print_output == 0:
            print("Iteration {k}/{m}, objective = {o}.".format(k=counter+1, \
                    m=no_of_iterations, o=objective_values[counter]))
    print("Iteration completed after {k}/{m}, objective = {o}.".format(k=counter + 1, \
                m=no_of_iterations, o=objective_values[counter]))
    return weights, objective_values
```

We implement the proximal map for the one-norm, also known as **soft_thresholding**. Our function takes a NumPy array *argument* and a scalar *threshold* as its arguments and performs the soft-thresholding operation to each component of the array.

```python
# Proximal map for lasso
def soft_thresholding(argument, threshold):
    return np.sign(argument) * np.maximum(0, np.abs(argument) - threshold)
```

Next, we write a function **lasso_cost_function** that implements the LASSO cost function as introduced in the lecture notes. The arguments include *data_matrix*, *weights* and *outputs* and a positive scalar *regularisation_parameter* that controls the balance between mean squared error and the one norm.

```python
# mse plus regularisation
def lasso_cost_function(data_matrix, weights, outputs, regularisation_parameter):
    return mean_squared_error(data_matrix, weights, outputs) + regularisation_parameter \
            * np.sum(np.abs(weights))
```

```python
# Lot of iterations to converge
initial_weights = np.zeros((lasso_matrix.shape[1], 1))
lasso_weights, lasso_objective_values = proximal_gradient_descent(objective, gradient, \
        proximal_map, initial_weights, step_size, no_of_iterations=300000, \
        print_output=30000)
```
