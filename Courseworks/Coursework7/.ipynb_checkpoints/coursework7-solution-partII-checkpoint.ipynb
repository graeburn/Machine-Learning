Ww write a function **nearest_neighbour_classification** that performs nearest neighbour classification as introduced in the lecture notes and that takes three arguments *inputs*, *dataset* and *no_of_neighbours*. The argument *inputs* contains a NumPy array of input samples to be classified. The argument *dataset* contains a list of data points for which we know the correct classification, and the argument *no_of_neighbours* specifies how many neighbours we take into account to determine the class labels for the inputs. Use the Euclidean norm to measure the distance between samples.

```python
def pairwise_distances(argument1, argument2):
    rows, cols = np.indices((argument1.shape[0], argument2.shape[0]))
    distances = np.sqrt(np.sum((argument1[rows, :] - argument2[cols, :])**2, axis=2))
    return distances

def tensor_broadcasting(argument1, argument2):
    return np.sqrt(np.sum((argument1[:,np.newaxis,:]-argument2[np.newaxis,:,:])**2, axis=2))

def nearest_neighbour_classification(inputs, dataset, no_of_neighbours):
    points, known_labels = dataset
    no_of_inputs = inputs.shape[0]
    no_of_points = points.shape[0]
    no_of_classes = 1 + np.max(known_labels)
    # Compute the pairwise distances
#    distances = pairwise_distances(inputs, points)
    # Alternative
    distances = tensor_broadcasting(inputs, points)
    # Sort indices and labels
    sorted_indices = np.argsort(distances, axis=1)
    sorted_labels = np.take_along_axis(np.broadcast_to(known_labels, (no_of_inputs, no_of_points)), \
                                          sorted_indices, 1)
    # Compute probabilities for each label based on labels of nearest neighbours
    rho = np.zeros((no_of_inputs, no_of_neighbours, no_of_classes))
    for counter in range(no_of_classes):
        rho[:, :, counter] = sorted_labels[:, :no_of_neighbours] == counter
    # Identify the correct label based on the previous probabilities
    recovered_labels = np.argmax(np.mean(rho, axis=1), axis=1)
    return recovered_labels

def classification_accuracy(estimated_labels, true_labels):
    equal_labels = estimated_labels == true_labels
    return np.mean(equal_labels)
```

We determine an optimal parameter *no_of_neighbours* based on $K$-fold cross validation and grid search.

```python
def grid_search(function, grid):
    no_grid_point = len(grid)
    function_values = []
    for counter in range(no_grid_point):
        function_values.append(function(grid[counter]))    
    return grid[np.argmax(function_values)], np.max(function_values)

def cross_validation(data, parameters, K):
    inputs, labels = data
    no_samples = len(labels)
    indices = np.random.permutation(no_samples)
    inputs = inputs[indices].reshape(K, -1, 2)
    labels = labels[indices].reshape(K, -1)
    list_of_optimal_neighbours = []
    classification_accuracies = []
    for counter in range(K):        
        validation_inputs = inputs[counter].reshape(-1, 2)        
        validation_labels = labels[counter].reshape(-1)                
        training_inputs = np.delete(inputs, counter, axis=0).reshape(-1, 2)
        training_labels = np.delete(labels, counter, axis=0).reshape(-1)
        training_data = [training_inputs, training_labels]
        classifier = lambda no_of_neighbours: classification_accuracy(nearest_neighbour_classification( \
                        validation_inputs, training_data, no_of_neighbours), validation_labels)
        optimal_no_of_neighbours, accuracy = grid_search(classifier, parameters)
        list_of_optimal_neighbours.append(optimal_no_of_neighbours)
        classification_accuracies.append(accuracy)
    return np.mean(optimal_no_of_neighbours), np.mean(classification_accuracies)
```

## Logistic regression

We get

$$ \frac{\partial (-\log \rho[y | X, W])}{\partial w_k} = \sum_{i=1}^s \text{softmax}(f(x_i, W))_k x_i - \sum_{i=1}^s 1_{y_i = k}x_i, $$

where $f$ is the linear model as specified in the lecture notes. Please note that for $f(X, W) = \Phi(X) W$ we can write the gradient in matrix form as

$$ \nabla_W -\log \rho[y | X, W]) = \Phi(X)^\top \left( \text{softmax}(\Phi(X) W) - Y \right) \, ,$$

where $Y \in \{0, 1\}^{s \times K}$ denotes the matrix that represents the one-hot vector encoding for each of the $s$ output labels $\{ y_i \}_{i = 1}^s$, assuming $K$ classes. Here, softmax is applied to the matrix $\Phi(X) W \in \mathbb{R}^{s \times K}$, acting on each row.

## Coordinate descent

Coordinate descent is an optimisation strategy where a function in multiple variables is minimised by minimising with respect to one coordinate at a time, whilst keeping the other coordinates fixed. Mathematically, this means we minimise a $1+d$-dimensional function $L$ by computing 
\begin{align}
w_l^{k + 1} = \arg\min_{w \in \mathbb{R}} L(w_0^{k + 1}, w_1^{k + 1}, \ldots, w_{l - 1}^{k + 1}, w, w_{l + 1}^k, \ldots, w_d^k )\label{eq:coordinate-descent}
\end{align}
for all $l \in \{0, 1, \ldots, d\}$ and $k = 1, 2, \ldots$.

First, we formulate the coordinate descent subproblem for the LASSO objective	
	\begin{align*}
	L(w_0, \ldots, w_d) = \frac{1}{2s} \sum_{i = 1}^s \left| \sum_{j = 0}^d \phi(x_i)_j w_j - y_i \right|^2 + \alpha \sum_{j = 0}^d | w_j |
	\end{align*}
as
    \begin{align*}
    w_l^{k + 1} = \arg\min_{w \in \mathbb{R}} \left\{ \frac{1}{2s} \sum_{i = 1}^s \left| \phi(x_i)_l w_l - \left( y_i - \sum_{j = 0}^{l - 1} \phi(x_i)_j w_j^{k + 1} - \sum_{j = l + 1}^d \phi(x_i)_j w_j^k \right) \right|^2 + \alpha | w_l | \right\} \, ,
    \end{align*}
	for $l \in \{0, 1, \ldots, d\}$, and derive the closed-form solution that reads
    \begin{align*}
    w_l^{k + 1} = \left( I + \frac{s \alpha}{\| \phi(x)_l \|^2} \partial | \cdot |\right)^{-1}\left( \frac{1}{\| \phi(x)_l \|^2}\left\langle \phi(x)_l \, , \, y_i - \sum_{j = 0}^{l - 1} \phi(x_i)_j w_j^{k + 1} - \sum_{j = l + 1}^d \phi(x_i)_j w_j^k\right\rangle \right)
    \end{align*}
    which we can verify in similar fashion as in Coursework 5. Next, we write a function that implements the solution to this problem.

```python
from coursework6 import soft_thresholding, mean_squared_error, lasso_cost_function

def lasso_coordinate_descent(data_matrix, initial_weights, outputs, regularisation_parameter, \
                             maximum_no_of_epochs=20, print_output=10):
    weights = np.copy(initial_weights)
    row_of_squared_norms = np.linalg.norm(data_matrix, axis=0) ** 2
    no_of_samples = data_matrix.shape[0]
    #indices = np.arange(data_matrix.shape[1])
    objective_values = []
    objective_values.append(lasso_cost_function(data_matrix, weights, outputs, regularisation_parameter))
    residual = outputs - data_matrix @ weights
    for counter in range(maximum_no_of_epochs):
        for index in range(data_matrix.shape[1]):            
            #residual = outputs - data_matrix[:, indices != index] @ weights[indices != index] 
            residual = residual + (data_matrix[:, index] * weights[index]).reshape(-1, 1)
            argument = (data_matrix[:, index] @ residual) / row_of_squared_norms[index]
            threshold = no_of_samples * regularisation_parameter / row_of_squared_norms[index]            
            weights[index] = soft_thresholding(argument, threshold)
            residual = residual - (data_matrix[:, index] * weights[index]).reshape(-1, 1)
        objective_values.append(lasso_cost_function(data_matrix, weights, outputs, regularisation_parameter))
        if (counter + 1) % print_output == 0:
            print("Epoch {k}/{m}, objective = {o}.".format(k=counter+1, \
                    m=maximum_no_of_epochs, o=objective_values[counter]))
    print("LASSO coordinate descent completed after {k}/{m} epochs, objective = {o}.".format(k=counter + 1, \
                m=maximum_no_of_epochs, o=objective_values[counter]))
    return weights, objective_values
```

Note that instead of evaluating 
\begin{align*}
y_i - \sum_{j = 0}^{l - 1} \phi(x_i)_j w_j^{k + 1} - \sum_{j = l + 1}^d \phi(x_i)_j w_j^k
\end{align*}
via matrix-vector multiplication at every iteration, we rather define 
\begin{align*}
\text{residual}^0 := y - \Phi(X) w^0 \, ,
\end{align*}
and update this quantity by computing
\begin{align*}
\text{residual}^{k - \frac{1}{2}}_l = \text{residual}^{k - 1}_l + \left( \begin{matrix} \phi(x_1)_l \\ \phi(x_2)_l \\ \vdots \\ \phi(x_s)_l \end{matrix} \right) w_l^{k - 1} \, ,
\end{align*}
subsequent computation of $w^k$ and modification of the residual via
\begin{align*}
\text{residual}^k_l = \text{residual}^{k - \frac{1}{2}}_l - \left( \begin{matrix} \phi(x_1)_l \\ \phi(x_2)_l \\ \vdots \\ \phi(x_s)_l \end{matrix} \right) w_l^k \, .
\end{align*}
This is computationally more efficient.

```python
from coursework6 import polynomial_basis
no_samples = 1000
sigma = 0.1
inputs = np.linspace(0, 1, no_samples)
true_outputs = np.sin(2*np.pi*inputs)
outputs = true_outputs + sigma * np.random.randn(no_samples)
degree = 20
lasso_matrix = polynomial_basis(inputs.reshape(-1, 1), degree)
regularisation_parameter = 1e-5
initial_weights = np.zeros((lasso_matrix.shape[1], 1))
```

We compute a solution numerically for a specified number of epochs via the following commands.

```python
maximum_no_of_epochs = 30000
lasso_weights, lasso_objective_values = lasso_coordinate_descent(lasso_matrix, initial_weights, \
                                        outputs.reshape(-1, 1), regularisation_parameter, \
                                        maximum_no_of_epochs, print_output=3000)
```

## Support vector machines

We conclude the coursework with an implementation of the dual maximisation problem for support vector machines with coordinate ascent. The dual SVM objective reads

\begin{align*}
L(\lambda_1, \ldots, \lambda_s) &= \langle \lambda, \textbf{1} \rangle - \frac{1}{2\alpha} \| \Phi(X)^\top Y \lambda \|^2 - \chi_{[0, 1]^s}(\lambda) \, , \\
&= \sum_{i = 1}^s \lambda_i - \frac{1}{2\alpha} \sum_{j = 0}^d \left| \sum_{i = 1}^s \phi(x_i)_j y_i \lambda_i \right|^2 - \sum_{i = 1}^s \chi_{[0, 1]}(\lambda_i) \, .
\end{align*}

Coordinate ascent aims at maximising $L$ coordinate-wise, i.e. 
\begin{align}
\lambda_l^{k + 1} = \arg\max_{\lambda \in \mathbb{R}} L(\lambda_1^{k + 1}, \lambda_2^{k + 1}, \ldots, \lambda_{l - 1}^{k + 1}, \lambda, \lambda_{l + 1}^k, \ldots, \lambda_s^k )
\end{align}
for all $l \in \{0, 1, \ldots, s\}$ and $k = 1, 2, \ldots$. For the dual SVM problem, each individual coordinate ascent update reads
\begin{align*}
\lambda_l^{k + 1} = \min\left(1, \max\left(0, \frac{1}{\| \phi(x_l) \|^2} \left( \alpha - \frac{\langle \phi(x_l), \sum_{i = 1}^{l - 1} \phi(x_i) y_i \lambda_i^{k + 1} + \sum_{i = l + 1}^{s} \phi(x_i) y_i \lambda_i^k \rangle }{y_l} \right) \right) \right) \, .
\end{align*}
We can implement this coordinate ascent strategy for the solution of the dual SVM problem with the following lines of codes.

```python
# Require multiple SVMs to use multiclass classification using an SVM (1To1 or 1ToRest approaches)

proximal_map = lambda argument: np.clip(argument, 0, 1)

def svm_dual_cost_function(data_matrix, lambdas, outputs, regularisation_parameter):
    return np.sum(lambdas) - 1/(2*regularisation_parameter) * (np.linalg.norm(data_matrix.T @ (np.diag(outputs) @ lambdas)) ** 2)

def svm_coordinate_ascent(data_matrix, outputs, regularisation_parameter, \
                             maximum_no_of_epochs=20, print_output=10):
    column_of_squared_norms = np.linalg.norm(data_matrix, axis=1) ** 2
    no_of_samples = data_matrix.shape[0]
    lambdas = np.zeros((no_of_samples, 1))    
    objective_values = []
    objective_values.append(svm_dual_cost_function(data_matrix, lambdas, outputs, regularisation_parameter))
    residual = data_matrix.T @ (np.diag(outputs) @ lambdas)
    for counter in range(maximum_no_of_epochs):
        for index in range(data_matrix.shape[0]):            
            residual = residual - (data_matrix[index, :] * outputs[index] * lambdas[index]).reshape(-1, 1)
            #argument = regularisation_parameter / column_of_squared_norms[index] - outputs[index] * (data_matrix[index, :] @ residual)
            argument = 1/column_of_squared_norms[index] * ( regularisation_parameter - (data_matrix[index, :] @ residual) / outputs[index] ) 
            lambdas[index] = proximal_map(argument)
            residual = residual + (data_matrix[index, :] * outputs[index] * lambdas[index]).reshape(-1, 1)
        objective_values.append(svm_dual_cost_function(data_matrix, lambdas, outputs, regularisation_parameter))
        if (counter + 1) % print_output == 0:
            print("Epoch {k}/{m}, objective = {o}.".format(k=counter+1, \
                    m=maximum_no_of_epochs, o=objective_values[counter]))
    print("SVM coordinate ascent completed after {k}/{m} epochs, objective = {o}.".format(k=counter + 1, \
                m=maximum_no_of_epochs, o=objective_values[counter]))
    return lambdas, objective_values
```

We compute a numerical approximation for an arbitrarily chosen regularisation parameter $\alpha$.

```python
regularisation_parameter = 1e+3
degree = 1
indices = np.random.permutation(len(height))
no_of_samples = 100
fewer_heights = 100*height[indices[:no_of_samples]] #scaling factor convert m to cm
fewer_weights = weight[indices[:no_of_samples]]
data_matrix = polynomial_basis(np.c_[fewer_heights.reshape(-1, 1), fewer_weights.reshape(-1, 1)], degree)
outputs = binary_sex[indices[:no_of_samples]]
outputs[binary_sex[indices[:no_of_samples]] == 0] = -1
maximum_no_of_epochs = 10000
lambdas, svm_dual_objective_values = svm_coordinate_ascent(data_matrix, outputs, \
                                        regularisation_parameter, maximum_no_of_epochs, print_output=1000)
```

Having computed (or approximated) $\hat \lambda$, we can compute the weights that solve the primal problem via $\hat w = (\Phi(X)^\top Y \hat \lambda)/\alpha$, which we do in the next line of code.

```python
svm_weights = 1/regularisation_parameter * ( (data_matrix.T @ np.diag(outputs)) @ lambdas )
print("The weights obtained with the SVM computed with dual coordinate ascent are {w}.T.".format(w=svm_weights.T))
```

We visualise the classification result obtained with the support vector machine.

```python
grid_matrix = polynomial_basis(grid_inputs, degree)
grid_svm_outputs = grid_matrix @ svm_weights
grid_svm_labels = (grid_svm_outputs > 0).astype(int) - (grid_svm_outputs < 0).astype(int)
plt.imshow(-grid_svm_labels.reshape(no_of_heights, no_of_weights), cmap='jet', extent=[130, 200, 20, 140], aspect="auto")
plt.plot(height[binary_sex==0]*100, weight[binary_sex==0], 'b.', linewidth=1)
plt.plot(height[binary_sex==1]*100, weight[binary_sex==1], 'r+', linewidth=1)
plt.xlabel('Height in metres', fontsize=16)
plt.xticks(fontsize=16)
plt.ylabel('Weight in kilogram', fontsize=16)
plt.yticks(fontsize=16)
plt.tight_layout;
```
