We begin by loading the NumPy and Matplotlib libraries and by setting the random seed to a fixed value.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(13) # Set seed so reproducable result. IMPORTANT TO MAKE RESULTS REPRODUCEABLE
```

We create our own dataset by defining a function $f(x) := \sin(2 \pi x)$ and by sampling the function at points $\{ x_i \}_{i = 1}^s$ from the interval $[0, 1]$. Subsequently, we compute the corresponding output samples $y_i^\dagger = f(x_i)$ for all $i \in \{1, \ldots, s\}$. In addition, we perturb our data with normal distributed random noise with mean zero and variance $\sigma^2$ in order to obtain noisy, simulated output data samples $\{ y_i \}_{i = 1}^s$. We refer to $s$ as *no_samples*.

```python
no_samples = 1000
sigma = 0.1
inputs = np.linspace(0, 1, no_samples)
true_outputs = np.sin(2*np.pi*inputs)
outputs = true_outputs + sigma * np.random.randn(no_samples) # np.random.randn computes samples based on normal distribution mean 0, variance 1
```

We visualise our data points.

```python
plt.plot(inputs, outputs, '--', linewidth=3, label='Noisy output samples')
plt.plot(inputs, true_outputs, linewidth=3, label='True output samples')
plt.legend( prop={'size': 12})
plt.xlabel('$x$-axis', fontsize=16)
plt.xticks(fontsize=16)
plt.ylabel('$y$-axis', fontsize=16)
plt.yticks(fontsize=16)
plt.tight_layout;
```

We split the previous dataset into two sets; one for testing, and one for cross-validation. The testing set contains 20\% randomly selected sample pairs, while the set for cross-validation contains the remaining 80\%.

```python
indices = np.random.permutation(no_samples)
no_indices_training = np.rint(0.8 * no_samples).astype(int)
training_indices, test_indices = np.sort(indices[:no_indices_training]), np.sort(indices[no_indices_training:])
training_data = [array[training_indices] for array in [inputs, outputs]]
test_data = [array[test_indices] for array in [inputs, outputs]]
```

Note how we have used list comprehension to apply identical operations to both inputs and outputs. We visualise both training and test data in the following plot.

```python
plt.plot(training_data[0], training_data[1], linewidth=3, label='Training data')
plt.plot(test_data[0], test_data[1], '--', linewidth=3, label='Test data')
plt.legend( prop={'size': 12})
plt.xlabel('$x$-axis', fontsize=16)
plt.xticks(fontsize=16)
plt.ylabel('$y$-axis', fontsize=16)
plt.yticks(fontsize=16)
plt.tight_layout;
```

Next, we Implement a grid-search strategy that evaluates the validation error 
\begin{align*}
\text{Val}_{S_v}(d) = \frac{1}{2|S_v|} \sum_{(x_i, y_i) \in S_v} | f_d(x_i) - y_i |^2 \, ,
\end{align*}
between output samples $\{ y_i^v \}_{i = 1}^{|S_v|}$ from a validation data set $S_v$ and the outputs of the polynomial regression function $f_d$ from last week's coursework for matching inputs $\{ x_i^v \}_{i = 1}^{|S_v|}$ from $S_v$. The parameter that is being evaluated on a grid is the degree $d$ of the polynomial. Here $|S_v|$ denotes the number of samples in the validation set $S_v$.

```python
def grid_search(function, grid):
    no_grid_point = len(grid)
    function_values = []
    for counter in range(no_grid_point):
        function_values.append(function(grid[counter]))    
    return grid[np.argmin(function_values)], np.min(function_values)

def polynomial_basis(inputs, degree=1):
    basis_matrix = np.ones((len(inputs), 1))
    for counter in range(1, degree + 1):
        basis_matrix = np.c_[basis_matrix, np.power(inputs, counter)]
    return basis_matrix

def polynomial_regression(basis_matrix, outputs):
    gram_matrix = basis_matrix.T @ basis_matrix
    right_hand_side = basis_matrix.T @ outputs.reshape(-1, 1)
    return np.linalg.solve(gram_matrix, right_hand_side)

def validate(training_samples, test_samples, degree):
    training_matrix = polynomial_basis(training_samples[0], degree)
    weights = polynomial_regression(training_matrix, training_samples[1])
    test_matrix = polynomial_basis(test_samples[0], degree)
    test_outputs = (test_matrix @ weights).reshape(-1)
    return validation_error(test_outputs, test_samples[1])

def validation_error(model_outputs, outputs):
    return np.mean((model_outputs - outputs) ** 2)/2
```

Now we combine the grid search strategy with a $K$-fold cross validation strategy as introduced in the lectures. We use $K$-fold cross validation by splitting our cross-validation dataset randomly into $K$ sets; we use $K-1$ sets to compute the polynomial regression weights (and consequently the function $f_d$) and the remaining set as the validation set $S_v$ for our grid search. We repeat this for all combinations, and average our results.

```python
def cross_validation(inputs, outputs, parameters, K):
    no_samples = len(inputs)
    indices = np.random.permutation(no_samples)
    inputs, outputs = [array[indices].reshape(-1, K) for array in [inputs, outputs]]
    data_samples = [inputs, outputs]
    optimal_degrees = []
    validation_errors = []
    for counter in range(K):        
        test_data = [array[:, counter] for array in data_samples]
        training_data = [np.delete(array, counter, axis=1).reshape(-1) for array in data_samples]
        validater = lambda degree: validate(training_data, test_data, degree)
        optimal_degree, validation_error = grid_search(validater, parameters)
        optimal_degrees.append(optimal_degree)
        validation_errors.append(validation_error)
    return np.mean(optimal_degrees), np.mean(validation_errors)
```

We define a grid of all degrees from $d=0$ to $d = 19$ and apply our combination of $K$-fold cross-validation and grid search for $K = 40$. Please feel free to experiment with different choices of $K$.

```python
degrees = np.arange(20)
K = 40
best_degree, best_validation_error = cross_validation(training_data[0], training_data[1], degrees, K)
print("The best degree on average for K = {k} is {d}.\nThe best validation error on average is {v}".format( \
        k = K, d = best_degree, v=best_validation_error))
```

We apply our model with degree $d$ for which we obtained the best results in the previous exercise to the test dataset.

```python
split_data = [[training_data[index], test_data[index]] for index in range(2)]
data_matrices = [polynomial_basis(data, np.rint(best_degree).astype(int)) for data in split_data[0]]
best_weights = polynomial_regression(data_matrices[0], training_data[1])
model_outputs = [matrix @ best_weights for matrix in data_matrices]
```

We visualise the true function $f$, the test samples and our optimal polynomial function $f_t$.

We see that by (approximately) minimising the validation error, we approximate the true function $f$ really well with our polynomial approximation $f_d$ for optimal degree $d$ (without the need to know $f$ in practice).
